Two variables x and y are selected at random. A dot is placed on the (x,y) coordinate. The density plot is plotted in rainbow colors. In this color scheme, the red regions correspond to those (x,y) values occur more often. 

```python
np.random.seed(0)
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.figure(figsize=(4,4))
sns.kdeplot(x=x, y=y, fill=True, cmap="rainbow", thresh=0, levels=100, alpha=0.9)
plt.title('Contour plot')
plt.show()
```

An object is observed to fall from a certain height. Here is the data of its position along the vertical axis (in m) as a function of time (s). The data is plotted and fitted to various functional forms. Why is such a fit important?

```python
y1 = np.array([0.171, 0.172, 0.231, 0.337, 0.466,0.6223,0.812])
t1 = np.array([0, 0.05,0.1,0.15,0.2,0.25,0.3])
def model_func2(t, m,b):
    return m*t + b

params, covariance = curve_fit(model_func2, t1, y1)

m_fit, b_fit = params
t_fit = np.linspace(min(t1),max(t1),50)
y_fit = model_func2(t_fit, m_fit, b_fit)
plt.scatter(t1,y1)
plt.plot(t_fit,y_fit)
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.show()
```

```python
def model_func3(t,a,v0):
    y0=0.17
    return y0+(v0*t)+(a*t**2/2)

params, covariance = curve_fit(model_func3, t1, y1)

a1_fit, v01_fit= params
print(a1_fit, v01_fit)
y_fit = model_func3(t_fit, a1_fit, v01_fit)
plt.scatter(t1,y1)
plt.plot(t_fit,y_fit)
plt.xlabel('time (s)')
plt.ylabel('height (m)')
```
